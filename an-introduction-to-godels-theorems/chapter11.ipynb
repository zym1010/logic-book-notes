# 11.8 Proving Q is order-adequate

Again, as in 11.1, just try to understand these proofs as we understand standard proofs in math.

There are some difference.

For example, to prove $\forall a \phi(a)$, you need to first prove $\phi(a)$ where $a$ is a free variable. Then use generalization rule (check [Mathematical Logic and Model Theory](https://doi.org/10.1007/978-1-4471-2176-3)) to add that $\forall$ symbol.

With this in mind, you should now understand the proof of (O1) better, as well as others better. $a + 0 = a$ is simply proved by axiom 4.

## logical equivalence -- a languge interpretation related concept

check https://en.wikipedia.org/wiki/Logical_equivalence. Basically this means that two sentences have the same value in each of possible all models (and under any given evaluation of free variables, I guess; check [Mathematical Logic and Model Theory](https://doi.org/10.1007/978-1-4471-2176-3) for what evaluation means).

> two statements are [logically] equivalent if they have the same truth value in every model (Mendelson 1979:56)
