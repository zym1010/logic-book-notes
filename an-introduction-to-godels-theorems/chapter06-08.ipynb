chapters 6, 7 prove two flavors of incompleteness thereom, as said in chapter 8, pp. 54

> Our arguments for incompleteness came in two flavours. First, we combined the premiss (a) that we are dealing with a sound theory with the premiss (b) that our theory's language is expressively rich enough. Second, we weakened one assumption and beefed up the other: in other words, we used the weaker premiss (a') that we are dealing with a consistent theory but added the stronger premiss (b') that our theory can prove enough facts. We'll see that Gödel’s arguments too come in these two flavours.

note that our therems are much weaker than Gödel's. Check chapter 8. But the spirit is the same.

## effectively axiomatizable wffs

Given in 6.3

Basically, a set of wffs are effectively axiomatized if it's the set of theoerems generated by a effectively axiomatized theory. Notice that "$\vdash$" is only defined on sentences. open wffs don't count. Check 4.4. So here "a set of wffs" is actually a set of sentences.
